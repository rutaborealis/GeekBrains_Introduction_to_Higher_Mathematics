```python
# Lesson 3
# Task 9.1
# Converting polar coordinates to a Cartesian coordinate system
import math

def polar_to_ccs(r, alpha):
    x = r * math.cos(alpha)
    y = r * math.sin(alpha)
    return(x, y)

print(polar_to_ccs(25, 90))
```

```python
import numpy
import matplotlib.pyplot as plot

plot.figure(figsize = (10, 10))

x = numpy.linspace(-5, 10, 1000)
y = (x - 2) ** 2 + 1
plot.plot(x, y, color='blue')

alpha = numpy.pi / 2
x1 = x * numpy.cos(alpha) + y * numpy.sin(alpha)
y1 = -x * numpy.sin(alpha) + y * numpy.cos(alpha)
plot.plot(x1, y1, color='red')
```
