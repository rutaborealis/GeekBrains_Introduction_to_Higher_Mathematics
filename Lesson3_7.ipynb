```python
# Lesson 3
# Task 7
import numpy
import matplotlib.pyplot as plot

plot.figure(figsize = (10, 10))

# x = numpy.linspace(-numpy.pi, numpy.pi, 500)
x=numpy.linspace(-10, 10, 1000)

# plot.plot(x, numpy.sin(x))
plot.plot(x, numpy.cos(x), color='blue')
plot.plot(x, 5 * numpy.cos(x - 10) + 5, color='green')
plot.plot(x, -5 * numpy.cos(x - 25) + 15, color='red')

plot.xlabel('x')
plot.ylabel('y')
plot.grid(True)
plot.show()
```
