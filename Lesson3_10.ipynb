```python
# Lesson 3
# Task 10.1
import numpy
import matplotlib.pyplot as plot
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    # print(x)
    # print(y)
    return (x ** 2 - 1 - y, numpy.exp(x) + x * (1 - y) - 1)

x = numpy.linspace(-2, 3, 500)

# print(equations([-1, 1]))
x1, y1 = fsolve(equations, (-1, 1))             
print (x1, y1)

plot.figure(figsize = (10, 10))
plot.plot(x, (numpy.exp(x) - 1) / x + 1, color='blue')
plot.plot(x, x ** 2 - 1, color='red')
plot.xlabel('x')
plot.ylabel('y')
plot.grid(True)
plot.ylim(-2, 8) 
```
