```python
# Lesson 3
# Task 3.1
import math
import numpy
import matplotlib.pyplot as plot

fig = plot.figure(figsize = (10, 10))

x = []
y1 = []
y2 = []
p = 100

for i in range(1000):
    x.append(i)
    y1.append(math.sqrt(2 * p * i))
    y2.append(-math.sqrt(2 * p * i))

plot.axes().set_aspect('equal')

plot.plot(x, y1, color='r')
plot.plot(x, y2, color='r')

plot.xlabel('x')
plot.ylabel('y')
```

```python
# Lesson 3
# Task 3.2
import math
import numpy
import matplotlib.pyplot as plot

fig = plot.figure(figsize = (10, 10))

x1 = []
x2 = []
y1 = []
y2 = []
r = 10000

for i in range(-r,r):
    x1.append(i)
    x2.append(-i)
    y1.append(math.sqrt(r ** 2 - i ** 2))
    y2.append(-math.sqrt(r ** 2 - i ** 2))

plot.axes().set_aspect('equal')

plot.plot(x1, y1, color='r')
plot.plot(x2, y2, color='r')

plot.xlabel('x')
plot.ylabel('y')
```

```python
# Lesson 3
# Task 3.3
import math
import numpy
import matplotlib.pyplot as plot

fig = plot.figure(figsize = (10, 10))

x1 = []
x2 = []
y1 = []
y2 = []
a = 10000
b = 5000

for i in range(-a,a):
    x1.append(i)
    x2.append(-i)
    y1.append(math.sqrt(b ** 2 - (b ** 2 * (i ** 2 / a ** 2))))
    y2.append(-math.sqrt(b ** 2 - (b ** 2 * (i ** 2 / a ** 2))))

plot.axes().set_aspect('equal')

plot.plot(x1, y1, color='r')
plot.plot(x2, y2, color='r')

plot.xlabel('x')
plot.ylabel('y')
```
