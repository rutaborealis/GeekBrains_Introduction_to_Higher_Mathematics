```python
# Lesson 5
# Task 2.2
import numpy
import matplotlib.pyplot as plot

n = 100
r = 0.7

x = numpy.random.rand(n)
y = r * x + (1 - r) * numpy.random.rand(n)

plot.plot(x, y, 'o')
plot.xlabel('x')
plot.ylabel('y')
plot.grid(True)
plot.show()

a = (numpy.sum(x) * numpy.sum(y) - n * numpy.sum(x * y )) / (numpy.sum(x) * numpy.sum(x) - n * numpy.sum(x * x))
b = (numpy.sum(y) - a * numpy.sum(x)) / n

A = numpy.vstack([x, numpy.ones(len(x))]).T
a1, b1 = numpy.linalg.lstsq(A, y, rcond=None)[0]
print(a, b)
print(a1, b1)

plot.plot([0, 1], [b, a + b])
plot.show()

c = numpy.corrcoef(x, y)
print('Коэффициент корреляции, рассчитанный автоматически: ', c)

x_avg = numpy.average(x)
y_avg = numpy.average(y)
print('Средние x, y: ', x_avg, y_avg)
r = numpy.sum((x - x_avg) * (y - y_avg)) / (numpy.sum((x - x_avg) ** 2) * (numpy.sum((y - y_avg) ** 2))) ** 0.5
print("Коэффициент корреляции, рассчитанный по формуле: ", r)
```
