```python
# Lesson 3
# Task 5.1
import math
import numpy
import matplotlib.pyplot as plot
from mpl_toolkits.mplot3d import Axes3D

#    
fig = plot.figure(figsize = (10, 10))

#     
ax = fig.add_subplot(1, 1, 1, projection='3d')

X = numpy.arange(-5, 5, 0.01)
Y = numpy.arange(-5, 5, 0.01)
X, Y = numpy.meshgrid(X, Y)
Z1 = 5 * X + 5 * Y
Z2 = 5 * X + 5 * Y + 300

ax.plot_surface(X, Y, Z1, color='blue')
ax.plot_wireframe(X, Y, Z2, colors='red')
```

```python
# Lesson 3
# Task 5.2
import math
import numpy
import matplotlib.pyplot as plot
from mpl_toolkits.mplot3d import Axes3D

#    
fig = plot.figure(figsize = (10, 10))

#     
ax = fig.add_subplot(1, 1, 1, projection='3d')

X = numpy.arange(-5, 5, 0.5)
Y = numpy.arange(-5, 5, 0.5)
X, Y = numpy.meshgrid(X, Y)
Z = X ** 2 + Y ** 2

ax.plot_surface(X, Y, Z, cmap=plot.cm.plasma)
```

```python
# Lesson 3
# Task 5.3
from mpl_toolkits.mplot3d import Axes3D
import numpy
import matplotlib.pyplot as plot

#  
f = lambda x, y: x ** 2 - y ** 2

#    
fig = plot.figure(figsize = (10, 10))

#     
ax = fig.add_subplot(1, 1, 1, projection='3d')

#     
xval = numpy.linspace(-5, 5, 1000)
yval = numpy.linspace(-5, 5, 1000)

#    xval   yval 
x, y = numpy.meshgrid(xval, yval)

#  z    x  y -       z
z = f(x, y)

#  
surf = ax.plot_surface(x, y, z, rstride=10, cstride=10, cmap=plot.cm.plasma)
#     
#    -   ,   -    
#   plasma
```

```python
# Lesson 3
# Task 5.4
from mpl_toolkits.mplot3d import Axes3D
import numpy
import matplotlib.pyplot as plot

#  
f = lambda x, y: numpy.sqrt(1 - (x ** 2) - (y ** 2))

#    
fig = plot.figure(figsize = (10, 10))

#     
ax = fig.add_subplot(1, 1, 1, projection='3d')

#     
xval = numpy.linspace(-1, 1, 5000)
yval = numpy.linspace(-1, 1, 5000)

#    xval   yval 
x, y = numpy.meshgrid(xval, yval)

#  z    x  y -       z
z = f(x, y)
z1 = -f(x, y)

#  
surf = ax.plot_surface(x, y, z, rstride=25, cstride=25, cmap=plot.cm.plasma)
surf1 = ax.plot_surface(x, y, z1, rstride=25, cstride=25, cmap=plot.cm.plasma)
#     
#    -   ,   -    
#   plasma
```
